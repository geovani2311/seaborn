## plots de regressão

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## regressão

```python
# tipos de marcadores
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=['x','v'], scatter_kws={'s': 100})
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", col="time", row="day")
```

```python
#tamanho
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=['x','v'], scatter_kws={'s': 100}, aspect=2.6)
```
