## plots matriciais

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
df = flights.pivot_table(index="year", columns="month", values="passengers")
```

##  heatmap - mapa contendo 3 eixos

```python
#tamanho + rotulo de dados + espessura das linhas

fig, ax = plt.subplots(figsize=(20,8))
sns.heatmap(df, ax=ax, annot=True, fmt=".0f", linewidths=1, cmap="magma")
```

```python
sns.clustermap(df)
```
