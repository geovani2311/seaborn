## plots categóricos

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### distribuição

```python
#passando funções para cacular
sns.barplot(x="sex", y="total_bill", data=tips)
```

```python
import numpy as np

#passando estimativa invés da média como no gráfico acima
sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.std)
```

```python
#contagem
sns.countplot(x="sex", data=tips)
```

## boxplot

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="rainbow")
```

```python
#horizontal
sns.boxplot(y="day", x="total_bill", data=tips, palette="rainbow", orient="h")
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="rainbow", hue="sex")
```

## violin

```python
sns.violinplot(x="day", y="total_bill", data=tips, palette="rainbow", hue="sex", split="True")
```

## swarmplot

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```
